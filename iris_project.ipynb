```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

iris=pd.read_csv("/content/sample_data/Iris.csv")
```

```python
print(iris.shape)
```

```python
print(iris.columns)
if "species" in iris.columns:
    species_counts = iris["species"].value_counts()
else:
    iris = iris.rename(columns={"Species": "species"})
    species_counts = iris["species"].value_counts()

print(species_counts)
```

```python
iris.plot(kind='scatter',x='SepalLengthCm',y='PetalLengthCm');
plt.show()
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(iris,hue="species")\
   .map(plt.scatter,"SepalLengthCm","SepalWidthCm") \
   .add_legend();
plt.show();
```

```python
iris['SepalLengthCm'].hist()
```

```python
iris['SepalWidthCm'].hist()
```

```python

import seaborn as sns
import matplotlib.pyplot as plt
# Boxplot
sns.boxplot(x="species", y="SepalLengthCm", data=iris)
plt.show()
sns.boxplot(x="species", y="SepalWidthCm", data=iris)
plt.show()
sns.boxplot(x="species", y="PetalLengthCm", data=iris)
plt.show()
sns.boxplot(x="species", y="PetalWidthCm", data=iris)
plt.show()

# Violinplot
sns.violinplot(x="species", y="SepalLengthCm", data=iris)
plt.show()
sns.violinplot(x="species", y="SepalWidthCm", data=iris)
plt.show()
sns.violinplot(x="species", y="PetalLengthCm", data=iris)
plt.show()
sns.violinplot(x="species", y="PetalWidthCm", data=iris)
plt.show()

# Pairplot
sns.pairplot(iris, hue="species")
plt.show()
```
